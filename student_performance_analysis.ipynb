```python
import pandas as pd

# creating dataset
data = {
    "Name": ["Mohit","Love","Palak","Aman","Riya"],
    "Marks": [96,69,96,75,88],
    "Roll_no": [11,23,24,25,26],
    "Branch": ["CSE","IT","CSE","ME","CSE"]
}

df = pd.DataFrame(data)

# saving and reading csv
df.to_csv("students.csv", index=False)
df = pd.read_csv("students.csv")

print(df)


# basic operations
df.sort_values("Marks", ascending=False).head(3)
df[df["Marks"] > 80]
df["Marks"].mean()


# intermediate
df.groupby("Branch")["Marks"].mean()
df.groupby("Branch").size()
df[(df["Branch"] == "CSE") & (df["Marks"] > 85)]


# advanced
def grade(m):
    if m >= 90:
        return "A"
    elif m >= 80:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(grade)

df.groupby("Grade").size()
df.groupby("Branch")["Marks"].agg(["max", "min", "mean"])
```
